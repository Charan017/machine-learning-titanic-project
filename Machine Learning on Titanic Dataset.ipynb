## CLASSIFICATION MODEL

### 1.LOGISTIC REGRESSION

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

#### DEFINING X &Y

```python
x=df.iloc[:,1:].values
x.shape
```

```python
y=df.iloc[:,0:1].values
y.shape
```

#### SPLITING THE DATASET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

#### BUILDING THE MODEL

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

#### Confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(y_test,y_pred)
print(confusion)
```

```python
TN=confusion[0,0]
FP=confusion[0,1]
FN=confusion[1,0]
TP=confusion[1,1]
```

```python
print(confusion)
print("TN:",TN)
print("FP:",FP)
print("FN:",FN)
print("TP:",TP)
```

```python
confusion_matrix=pd.DataFrame(confusion)
confusion_matrix.columns=['predicted not survived','predicted survived']
confusion_matrix=confusion_matrix.rename(index={0:'Actual not survived',1:'Actual survived'})
confusion_matrix
```

#### Confusion Matrix Evalution Metrics

#### Classification Accuracy

```python
from sklearn import metrics
accuracy=metrics.accuracy_score(y_test,y_pred)
accuracy1=(TN+TP)/(TN+FP+FN+TP)
print("accuracy from metrics:",accuracy)
print("accuracy1 from metrics:",accuracy1)
```

#### Classification Error

```python
print((FP+FN)/float(TP+FP+FN+TN))
print(round(1-metrics.accuracy_score(y_test,y_pred),4))
```

#### Sensitivity

```python
print("RECALL:",TP/(TP+FN))
```

#### Specificity

```python
print("SPECIFICITY:",TN/(TN+FP))
```

#### FALSE POSITIVE RATE

```python
print("FALSE POSITIVE RATE: ",FN/(FN+TP))
```

#### FALSE NEGATIVE RATE

```python
print("FALSE NEGATIVE RATE: ",FP/(FP+TN))
```

#### PRECISION

```python
print("precision:",round(TP/float(TP+FP),2))
```

#### f1score=2pr/(p+r)

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,roc_auc_score
accuracy=accuracy_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
precision=precision_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
roc_auc=roc_auc_score(y_test,y_pred)

print("Accuracy is:",round(accuracy,2)*100)
print("f1 is:",round(f1,2)*100)
print("Precision is:",round(precision,2)*100)
print("recall is:",round(recall,2)*100)
print("roc_auc is:",round(roc_auc,2)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
