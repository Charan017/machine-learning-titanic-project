### Step 5:Outliers

#### Age

```python
df.boxplot(column=['Age'])
```

```python
df['Age'].describe()
```

```python
IQR=df['Age'].quantile(0.75)-df['Age'].quantile(0.25)
print(IQR)
```

```python
Upper_OutlierLimit=df['Age'].quantile(0.75)+1.5*IQR
print(Upper_OutlierLimit)
```

```python
Upper_OutlierValues=df[(df['Age']>=Upper_OutlierLimit)]
```

```python
Upper_OutlierValues
```

```python
li=[Upper_OutlierValues.iloc[0:100,0]-1]
```

```python
li
```

```python
fill=df['Age'].quantile(0.95)
for x in li:
    df.loc[x,'Age']=fill
```

```python
df.boxplot(column=['Age'])
```

```python
Lower_OutlierLimit=df['Age'].quantile(0.25)-1.5*IQR
print(Lower_OutlierLimit)
```

```python
Lower_OutlierValues=df[(df['Age']<=Lower_OutlierLimit)]
```

```python
Lower_OutlierValues
```

```python
li=[Lower_OutlierValues.iloc[0:100,0]-1]
```

```python
li
```

```python
ab=df['Age'].quantile(0.95)
for x in li:
    df.loc[x,'Age']=ab
```

```python
df.boxplot(column=['Age'])
```
